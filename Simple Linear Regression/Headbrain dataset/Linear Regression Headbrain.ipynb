```python
# First step is to import libraries which are used for Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset and view it 
dataset=pd.read_csv('headbrain.csv')
dataset
```

```python
x=dataset.iloc[:,2:3].values
x
```

```python
# Separation of features and target (dependent and independent values)
y=dataset.iloc[:,-1].values
y
```

```python
# Select the model for training and testing using train_test_split library

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
# Viewing all the individual datas by plotting using matplotlib library
plt.scatter(x_train,y_train)
```

```python
# Linear regression line is drawn using R sqaured value. 
#The smaller the R value, the greater is the minimisation of error
# minimus R square value indicates the best fit line in linear regression

y_pred=reg.predict(x_train)
y_pred
plt.plot(x_train,y_pred,'r')
```
